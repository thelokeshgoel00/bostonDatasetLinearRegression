```python
def hypothesis(x,theta):
    y_=0.0
    n= x.shape[0]
    for i in range (n):
        y_+=(theta[i]*x[i])
    return y_


def error(X,y,theta):
    e=0.0
    m=X.shape[0]
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        e+= (y[i]-y_)**2
    return e/m    
    
    
def gradient(X,y,theta):
    m,n=X.shape
    grad= np.zeros((n,))
    for j in range(n):
        for i in range(m):
            y_=hypothesis(X[i],theta)
            grad[j]+=(y_-y[i])*X[i][j]
    return grad/m        
            
    
    
def gradient_descent(X,y,learning_rate=0.1,max_epochs=300):
    m,n=X.shape
    theta=np.zeros((n,))
    error_list=[]
    
    for i in range(max_epochs):
        e= error(X,y,theta)
        error_list.append(e)
        grad = gradient(X,y,theta)
        
        for j in range(n):
            theta[j]=theta[j]-learning_rate*grad[j]
    return theta,error_list
```

```python
import time
start= time.time()
theta,error_list=gradient_descent(X,Y)
end= time.time()
print("time taken is", end-start)
```
